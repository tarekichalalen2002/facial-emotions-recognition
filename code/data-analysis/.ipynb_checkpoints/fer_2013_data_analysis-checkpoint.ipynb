```python
# FER2013 analysis: imports and setup
import os
import random
import math
from typing import List, Dict, Tuple, Optional

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import datasets as tv_datasets

# Style and reproducibility
plt.style.use("seaborn-v0_8-whitegrid")
random.seed(42)
np.random.seed(42)

# Output directory (current working directory)
OUTPUT_DIR = os.getcwd()
print(f"Plots will be saved to: {OUTPUT_DIR}")

# Common image extensions
IMAGE_EXTENSIONS = {".jpg", ".jpeg", ".png", ".bmp", ".tiff"}
```

```python
# Locate FER2013 root and load train/test splits
# Try both the user-specified path (../datsets/FER_2013) and a likely corrected fallback (../datasets/FER_2013)
CANDIDATE_FER_ROOTS = [
    os.path.abspath(os.path.join(os.getcwd(), "../datsets/FER_2013")),
    os.path.abspath(os.path.join(os.getcwd(), "../datasets/FER_2013")),
]

fer_root = None
for p in CANDIDATE_FER_ROOTS:
    if os.path.isdir(p):
        fer_root = p
        break

if fer_root is None:
    print("FER2013 root not found. Checked:")
    for p in CANDIDATE_FER_ROOTS:
        print(" -", p)
else:
    print("Using FER2013 root:", fer_root)

fer_train_dataset = None
fer_test_dataset = None

if fer_root is not None:
    fer_train_dir = os.path.join(fer_root, "train")
    fer_test_dir = os.path.join(fer_root, "test")

    if os.path.isdir(fer_train_dir):
        fer_train_dataset = tv_datasets.ImageFolder(fer_train_dir)
        print(f"Loaded FER2013 train split from {fer_train_dir} with {len(fer_train_dataset)} images and classes: {fer_train_dataset.classes}")
    else:
        print("Train directory not found:", fer_train_dir)

    if os.path.isdir(fer_test_dir):
        fer_test_dataset = tv_datasets.ImageFolder(fer_test_dir)
        print(f"Loaded FER2013 test split from {fer_test_dir} with {len(fer_test_dataset)} images and classes: {fer_test_dataset.classes}")
    else:
        print("Test directory not found:", fer_test_dir)
```

```python
# Helpers: save figures and image scanning
from collections import Counter


def save_figure(fig, filename: str):
    fig.tight_layout()
    out_path = os.path.join(OUTPUT_DIR, filename)
    fig.savefig(out_path, dpi=150, bbox_inches="tight")
    plt.close(fig)


def scan_image_sizes_recursive(root_dir: str, sample_size: Optional[int] = 5000) -> Dict[str, List[int]]:
    if not root_dir or not os.path.isdir(root_dir):
        return {"widths": [], "heights": [], "aspect_ratios": [], "areas": []}

    # Collect image file paths recursively
    image_paths = []
    for subdir, _dirs, files in os.walk(root_dir):
        for f in files:
            ext = os.path.splitext(f)[1].lower()
            if ext in IMAGE_EXTENSIONS:
                image_paths.append(os.path.join(subdir, f))

    if len(image_paths) == 0:
        return {"widths": [], "heights": [], "aspect_ratios": [], "areas": []}

    if sample_size is not None and len(image_paths) > sample_size:
        image_paths = random.sample(image_paths, sample_size)

    widths, heights, aspects, areas = [], [], [], []

    for fp in image_paths:
        try:
            with Image.open(fp) as im:
                w, h = im.size
        except Exception:
            continue
        if w <= 0 or h <= 0:
            continue
        widths.append(w)
        heights.append(h)
        aspects.append(w / h)
        areas.append(w * h)

    return {"widths": widths, "heights": heights, "aspect_ratios": aspects, "areas": areas}


def plot_image_stats(stats: Dict[str, List[int]], split_name: str):
    widths = stats.get("widths", [])
    heights = stats.get("heights", [])
    aspects = stats.get("aspect_ratios", [])
    areas = stats.get("areas", [])

    if len(widths) > 0:
        fig, ax = plt.subplots(figsize=(6, 4))
        ax.hist(widths, bins=50, color="#4C78A8")
        ax.set_title(f"Image widths ({split_name})")
        ax.set_xlabel("Width (px)")
        ax.set_ylabel("Count")
        save_figure(fig, f"image_width_hist_{split_name}.png")

    if len(heights) > 0:
        fig, ax = plt.subplots(figsize=(6, 4))
        ax.hist(heights, bins=50, color="#54A24B")
        ax.set_title(f"Image heights ({split_name})")
        ax.set_xlabel("Height (px)")
        ax.set_ylabel("Count")
        save_figure(fig, f"image_height_hist_{split_name}.png")

    if len(aspects) > 0:
        fig, ax = plt.subplots(figsize=(6, 4))
        ax.hist(aspects, bins=50, color="#E45756")
        ax.set_title(f"Aspect ratios (W/H) ({split_name})")
        ax.set_xlabel("Aspect ratio")
        ax.set_ylabel("Count")
        save_figure(fig, f"image_aspect_ratio_hist_{split_name}.png")

    if len(widths) > 0 and len(heights) > 0:
        fig, ax = plt.subplots(figsize=(5.5, 5.5))
        ax.hexbin(widths, heights, gridsize=40, cmap="magma", mincnt=1)
        ax.set_title(f"Width vs Height (hexbin) ({split_name})")
        ax.set_xlabel("Width (px)")
        ax.set_ylabel("Height (px)")
        save_figure(fig, f"image_width_height_hexbin_{split_name}.png")

        fig, ax = plt.subplots(figsize=(5.5, 5.5))
        ax.scatter(widths, heights, s=4, alpha=0.2, color="#72B7B2")
        ax.set_title(f"Width vs Height (scatter) ({split_name})")
        ax.set_xlabel("Width (px)")
        ax.set_ylabel("Height (px)")
        save_figure(fig, f"image_width_height_scatter_{split_name}.png")

    if len(areas) > 0:
        fig, ax = plt.subplots(figsize=(6, 4))
        ax.hist(areas, bins=50, color="#F58518")
        ax.set_title(f"Image areas ({split_name})")
        ax.set_xlabel("Area (px^2)")
        ax.set_ylabel("Count")
        ax.set_xscale("log")
        save_figure(fig, f"image_area_hist_{split_name}.png")
```

```python
# Class distribution analysis and plots for FER2013 train/test

def plot_class_distribution(imagefolder_ds: tv_datasets.ImageFolder, split_name: str):
    # Count items per class by iterating dataset targets
    if imagefolder_ds is None:
        return
    class_names = imagefolder_ds.classes
    num_classes = len(class_names)

    counts = np.zeros(num_classes, dtype=int)
    for _path, target in imagefolder_ds.samples:
        if 0 <= target < num_classes:
            counts[target] += 1

    total = counts.sum()
    print(f"\n--- {split_name.capitalize()} Distribution ---")
    for name, val in zip(class_names, counts):
        pct = (val / total * 100.0) if total > 0 else 0.0
        print(f"{name:12s} : {val:6d} ({pct:5.2f}%)")

    # Pie chart
    fig, ax = plt.subplots(figsize=(6, 6))
    colors = plt.cm.tab20(np.linspace(0, 1, num_classes))
    ax.pie(counts, labels=class_names, autopct='%1.1f%%', startangle=90, colors=colors, pctdistance=0.75)
    ax.axis('equal')
    ax.set_title(f'{split_name.capitalize()} expression distribution (FER2013)')
    save_figure(fig, f'expression_distribution_{split_name}_fer.png')


# Run distribution plots for train and test
if fer_train_dataset is not None:
    plot_class_distribution(fer_train_dataset, 'train')
else:
    print('Skipping train distribution: dataset not loaded.')

if fer_test_dataset is not None:
    plot_class_distribution(fer_test_dataset, 'test')
else:
    print('Skipping test distribution: dataset not loaded.')
```

```python
# Image size statistics: scan recursively inside train and test folders

if fer_root is not None:
    # Train image stats
    train_dir = os.path.join(fer_root, 'train')
    train_stats = scan_image_sizes_recursive(train_dir, sample_size=5000)
    plot_image_stats(train_stats, 'train_fer')
    print('Saved image stat plots for FER2013 train.')

    # Test image stats
    test_dir = os.path.join(fer_root, 'test')
    test_stats = scan_image_sizes_recursive(test_dir, sample_size=5000)
    plot_image_stats(test_stats, 'test_fer')
    print('Saved image stat plots for FER2013 test.')
else:
    print('Skipping image size statistics: FER2013 root not found.')
```

```python
# Display and save one example image per class from FER2013 train split (fallback to test)

def show_one_example_per_class(imagefolder_ds: Optional[tv_datasets.ImageFolder], title_prefix: str):
    if imagefolder_ds is None:
        return False
    class_to_example_path: Dict[int, str] = {}

    # Build mapping of class index to a sample path
    for path, target in imagefolder_ds.samples:
        if target not in class_to_example_path:
            class_to_example_path[target] = path
        if len(class_to_example_path) == len(imagefolder_ds.classes):
            break

    if len(class_to_example_path) == 0:
        return False

    ordered_indices = sorted(class_to_example_path.keys())
    cols = 4
    rows = math.ceil(len(ordered_indices) / cols)
    fig, axes = plt.subplots(rows, cols, figsize=(cols * 3.2, rows * 3.2))

    if rows == 1 and cols == 1:
        axes = np.array([[axes]])
    elif rows == 1:
        axes = np.array([axes])
    elif cols == 1:
        axes = axes.reshape(rows, 1)

    for idx, class_idx in enumerate(ordered_indices):
        r = idx // cols
        c = idx % cols
        ax = axes[r, c]
        img_path = class_to_example_path[class_idx]
        try:
            with Image.open(img_path) as im:
                ax.imshow(im)
        except Exception:
            ax.text(0.5, 0.5, 'Load error', ha='center', va='center')
        ax.set_title(imagefolder_ds.classes[class_idx])
        ax.axis('off')

    # Hide any unused axes
    for idx in range(len(ordered_indices), rows * cols):
        r = idx // cols
        c = idx % cols
        axes[r, c].axis('off')

    fig.suptitle(f'{title_prefix}: One example per class', y=1.02)
    fig.tight_layout()
    out_path = os.path.join(OUTPUT_DIR, f'fer_examples_{title_prefix.lower()}.png')
    fig.savefig(out_path, dpi=150, bbox_inches='tight')
    plt.close(fig)
    print(f'Saved: {out_path}')
    return True

if not show_one_example_per_class(fer_train_dataset, 'FER2013 Train'):
    show_one_example_per_class(fer_test_dataset, 'FER2013 Test')
```

```python
# Combined train vs test class distribution (two pie charts)

def build_class_counts(imagefolder_ds: Optional[tv_datasets.ImageFolder]) -> Dict[str, int]:
    if imagefolder_ds is None:
        return {}
    counts: Dict[str, int] = {name: 0 for name in imagefolder_ds.classes}
    for _path, target in imagefolder_ds.samples:
        if 0 <= target < len(imagefolder_ds.classes):
            name = imagefolder_ds.classes[target]
            counts[name] += 1
    return counts

if fer_train_dataset is not None and fer_test_dataset is not None:
    train_counts = build_class_counts(fer_train_dataset)
    test_counts = build_class_counts(fer_test_dataset)

    all_classes = sorted(set(train_counts.keys()) | set(test_counts.keys()))
    train_vals = [train_counts.get(c, 0) for c in all_classes]
    test_vals = [test_counts.get(c, 0) for c in all_classes]

    # Print summary
    print("\n--- FER2013 Train Distribution ---")
    total_train = sum(train_vals)
    for c, v in zip(all_classes, train_vals):
        pct = (v / total_train * 100.0) if total_train > 0 else 0.0
        print(f"{c:12s} : {v:6d} ({pct:5.2f}%)")

    print("\n--- FER2013 Test Distribution ---")
    total_test = sum(test_vals)
    for c, v in zip(all_classes, test_vals):
        pct = (v / total_test * 100.0) if total_test > 0 else 0.0
        print(f"{c:12s} : {v:6d} ({pct:5.2f}%)")

    colors = plt.cm.tab20(np.linspace(0, 1, len(all_classes)))
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    axes[0].pie(train_vals, labels=all_classes, autopct='%1.1f%%', startangle=90, colors=colors, pctdistance=0.75)
    axes[0].axis('equal')
    axes[0].set_title('FER2013 Train expression distribution')

    axes[1].pie(test_vals, labels=all_classes, autopct='%1.1f%%', startangle=90, colors=colors, pctdistance=0.75)
    axes[1].axis('equal')
    axes[1].set_title('FER2013 Test expression distribution')

    save_figure(fig, 'expression_distribution_train_vs_test_fer.png')
    print('Saved combined train vs test expression distribution plot.')
else:
    print('Skipping combined train vs test distribution: one or both splits not loaded.')
```
