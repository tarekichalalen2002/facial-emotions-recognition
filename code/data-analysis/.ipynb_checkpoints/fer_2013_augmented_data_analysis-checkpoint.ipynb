```python
import torch
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, Dataset,random_split
import matplotlib.pyplot as plt
import torch.optim as optim
import numpy as np
from timm.models.vision_transformer import VisionTransformer
import timm
import os
from tqdm import tqdm
from PIL import Image
from torch.utils.data import ConcatDataset
import torch
from torchvision import datasets
import timm
import copy
import torch.nn as nn
import torch.optim as optim
import math
```

```python
train_dir = "../datasets/FER_2013_enhanced/FER_preprocessed_train"
test_dir = "../datasets/FER_2013_enhanced/preprocessed_test"
train_transform = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),  # random crop & scale
    transforms.RandomHorizontalFlip(p=0.5),               # flip horizontally
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),  # color augmentation
    transforms.RandomRotation(15),                        # small rotations
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225]),
])
val_test_transform = transforms.Compose([
    transforms.Resize((224, 224)),  # deterministic resize for val/test
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225]),
])

full_train_dataset = datasets.ImageFolder(root=train_dir, transform=train_transform)
test_dataset = datasets.ImageFolder(root=test_dir, transform=val_test_transform)
```

```python
# Analysis setup and helpers (mirrors fer_2013_data_analysis)
import random
from typing import Optional, Dict, List

# Style and reproducibility
plt.style.use("seaborn-v0_8-whitegrid")
random.seed(42)
np.random.seed(42)

# Output directory (current working directory)
OUTPUT_DIR = os.getcwd()
print(f"Plots will be saved to: {OUTPUT_DIR}")

# Common image extensions
IMAGE_EXTENSIONS = {".jpg", ".jpeg", ".png", ".bmp", ".tiff"}


def save_figure(fig, filename: str):
    fig.tight_layout()
    out_path = os.path.join(OUTPUT_DIR, filename)
    fig.savefig(out_path, dpi=150, bbox_inches="tight")
    plt.close(fig)


def scan_image_sizes_recursive(root_dir: str, sample_size: Optional[int] = 5000) -> Dict[str, List[int]]:
    if not root_dir or not os.path.isdir(root_dir):
        return {"widths": [], "heights": [], "aspect_ratios": [], "areas": []}

    # Collect image file paths recursively
    image_paths = []
    for subdir, _dirs, files in os.walk(root_dir):
        for f in files:
            ext = os.path.splitext(f)[1].lower()
            if ext in IMAGE_EXTENSIONS:
                image_paths.append(os.path.join(subdir, f))

    if len(image_paths) == 0:
        return {"widths": [], "heights": [], "aspect_ratios": [], "areas": []}

    if sample_size is not None and len(image_paths) > sample_size:
        image_paths = random.sample(image_paths, sample_size)

    widths, heights, aspects, areas = [], [], [], []

    for fp in image_paths:
        try:
            with Image.open(fp) as im:
                w, h = im.size
        except Exception:
            continue
        if w <= 0 or h <= 0:
            continue
        widths.append(w)
        heights.append(h)
        aspects.append(w / h)
        areas.append(w * h)

    return {"widths": widths, "heights": heights, "aspect_ratios": aspects, "areas": areas}


def plot_image_stats(stats: Dict[str, List[int]], split_name: str):
    widths = stats.get("widths", [])
    heights = stats.get("heights", [])
    aspects = stats.get("aspect_ratios", [])
    areas = stats.get("areas", [])

    if len(widths) > 0:
        fig, ax = plt.subplots(figsize=(6, 4))
        ax.hist(widths, bins=50, color="#4C78A8")
        ax.set_title(f"Image widths ({split_name})")
        ax.set_xlabel("Width (px)")
        ax.set_ylabel("Count")
        save_figure(fig, f"image_width_hist_{split_name}.png")

    if len(heights) > 0:
        fig, ax = plt.subplots(figsize=(6, 4))
        ax.hist(heights, bins=50, color="#54A24B")
        ax.set_title(f"Image heights ({split_name})")
        ax.set_xlabel("Height (px)")
        ax.set_ylabel("Count")
        save_figure(fig, f"image_height_hist_{split_name}.png")

    if len(aspects) > 0:
        fig, ax = plt.subplots(figsize=(6, 4))
        ax.hist(aspects, bins=50, color="#E45756")
        ax.set_title(f"Aspect ratios (W/H) ({split_name})")
        ax.set_xlabel("Aspect ratio")
        ax.set_ylabel("Count")
        save_figure(fig, f"image_aspect_ratio_hist_{split_name}.png")

    if len(widths) > 0 and len(heights) > 0:
        fig, ax = plt.subplots(figsize=(5.5, 5.5))
        ax.hexbin(widths, heights, gridsize=40, cmap="magma", mincnt=1)
        ax.set_title(f"Width vs Height (hexbin) ({split_name})")
        ax.set_xlabel("Width (px)")
        ax.set_ylabel("Height (px)")
        save_figure(fig, f"image_width_height_hexbin_{split_name}.png")

        fig, ax = plt.subplots(figsize=(5.5, 5.5))
        ax.scatter(widths, heights, s=4, alpha=0.2, color="#72B7B2")
        ax.set_title(f"Width vs Height (scatter) ({split_name})")
        ax.set_xlabel("Width (px)")
        ax.set_ylabel("Height (px)")
        save_figure(fig, f"image_width_height_scatter_{split_name}.png")

    if len(areas) > 0:
        fig, ax = plt.subplots(figsize=(6, 4))
        ax.hist(areas, bins=50, color="#F58518")
        ax.set_title(f"Image areas ({split_name})")
        ax.set_xlabel("Area (px^2)")
        ax.set_ylabel("Count")
        ax.set_xscale("log")
        save_figure(fig, f"image_area_hist_{split_name}.png")
```

```python
# Class distribution analysis and pies for augmented FER2013 train/test

def plot_class_distribution_aug(imagefolder_ds, split_name: str):
    if imagefolder_ds is None:
        return
    class_names = imagefolder_ds.classes
    num_classes = len(class_names)

    counts = np.zeros(num_classes, dtype=int)
    for _path, target in imagefolder_ds.samples:
        if 0 <= target < num_classes:
            counts[target] += 1

    total = counts.sum()
    print(f"\n--- {split_name.capitalize()} (AUG) Distribution ---")
    for name, val in zip(class_names, counts):
        pct = (val / total * 100.0) if total > 0 else 0.0
        print(f"{name:12s} : {val:6d} ({pct:5.2f}%)")

    # Pie chart
    fig, ax = plt.subplots(figsize=(6, 6))
    colors = plt.cm.tab20(np.linspace(0, 1, num_classes))
    ax.pie(counts, labels=class_names, autopct='%1.1f%%', startangle=90, colors=colors, pctdistance=0.75)
    ax.axis('equal')
    ax.set_title(f'{split_name.capitalize()} expression distribution (FER2013 Augmented)')
    save_figure(fig, f'expression_distribution_{split_name}_fer_aug.png')


# Run distribution pies for train and test
if 'full_train_dataset' in globals() and full_train_dataset is not None:
    plot_class_distribution_aug(full_train_dataset, 'train')
else:
    print('Skipping train distribution: full_train_dataset not loaded.')

if 'test_dataset' in globals() and test_dataset is not None:
    plot_class_distribution_aug(test_dataset, 'test')
else:
    print('Skipping test distribution: test_dataset not loaded.')
```

```python
# Image size statistics for augmented train/test

if os.path.isdir(train_dir):
    aug_train_stats = scan_image_sizes_recursive(train_dir, sample_size=5000)
    plot_image_stats(aug_train_stats, 'train_fer_aug')
    print('Saved image stat plots for FER2013 augmented train.')
else:
    print('Skipping augmented train image stats: train_dir not found.')

if os.path.isdir(test_dir):
    aug_test_stats = scan_image_sizes_recursive(test_dir, sample_size=5000)
    plot_image_stats(aug_test_stats, 'test_fer_aug')
    print('Saved image stat plots for FER2013 augmented test.')
else:
    print('Skipping augmented test image stats: test_dir not found.')
```

```python
# Show one example image per class from augmented train (fallback to test)

def show_one_example_per_class(imagefolder_ds, title_prefix: str):
    if imagefolder_ds is None:
        return False
    class_to_example_path: Dict[int, str] = {}

    for path, target in imagefolder_ds.samples:
        if target not in class_to_example_path:
            class_to_example_path[target] = path
        if len(class_to_example_path) == len(imagefolder_ds.classes):
            break

    if len(class_to_example_path) == 0:
        return False

    ordered_indices = sorted(class_to_example_path.keys())
    cols = 4
    rows = math.ceil(len(ordered_indices) / cols)
    fig, axes = plt.subplots(rows, cols, figsize=(cols * 3.2, rows * 3.2))

    if rows == 1 and cols == 1:
        axes = np.array([[axes]])
    elif rows == 1:
        axes = np.array([axes])
    elif cols == 1:
        axes = axes.reshape(rows, 1)

    for idx, class_idx in enumerate(ordered_indices):
        r = idx // cols
        c = idx % cols
        ax = axes[r, c]
        img_path = class_to_example_path[class_idx]
        try:
            with Image.open(img_path) as im:
                ax.imshow(im)
        except Exception:
            ax.text(0.5, 0.5, 'Load error', ha='center', va='center')
        ax.set_title(imagefolder_ds.classes[class_idx])
        ax.axis('off')

    for idx in range(len(ordered_indices), rows * cols):
        r = idx // cols
        c = idx % cols
        axes[r, c].axis('off')

    fig.suptitle(f'{title_prefix}: One example per class', y=1.02)
    fig.tight_layout()
    out_path = os.path.join(OUTPUT_DIR, f'fer_examples_{title_prefix.lower().replace(" ", "_")}.png')
    fig.savefig(out_path, dpi=150, bbox_inches='tight')
    plt.close(fig)
    print(f'Saved: {out_path}')
    return True

if 'full_train_dataset' in globals() and not show_one_example_per_class(full_train_dataset, 'FER2013 Augmented Train'):
    show_one_example_per_class(test_dataset, 'FER2013 Augmented Test')
```

```python
# Combined train vs test class distribution pies (augmented)

def build_class_counts(imagefolder_ds) -> Dict[str, int]:
    if imagefolder_ds is None:
        return {}
    counts: Dict[str, int] = {name: 0 for name in imagefolder_ds.classes}
    for _path, target in imagefolder_ds.samples:
        if 0 <= target < len(imagefolder_ds.classes):
            name = imagefolder_ds.classes[target]
            counts[name] += 1
    return counts

if 'full_train_dataset' in globals() and full_train_dataset is not None and 'test_dataset' in globals() and test_dataset is not None:
    train_counts = build_class_counts(full_train_dataset)
    test_counts = build_class_counts(test_dataset)

    all_classes = sorted(set(train_counts.keys()) | set(test_counts.keys()))
    train_vals = [train_counts.get(c, 0) for c in all_classes]
    test_vals = [test_counts.get(c, 0) for c in all_classes]

    # Print summary
    print("\n--- FER2013 Augmented Train Distribution ---")
    total_train = sum(train_vals)
    for c, v in zip(all_classes, train_vals):
        pct = (v / total_train * 100.0) if total_train > 0 else 0.0
        print(f"{c:12s} : {v:6d} ({pct:5.2f}%)")

    print("\n--- FER2013 Augmented Test Distribution ---")
    total_test = sum(test_vals)
    for c, v in zip(all_classes, test_vals):
        pct = (v / total_test * 100.0) if total_test > 0 else 0.0
        print(f"{c:12s} : {v:6d} ({pct:5.2f}%)")

    colors = plt.cm.tab20(np.linspace(0, 1, len(all_classes)))
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    axes[0].pie(train_vals, labels=all_classes, autopct='%1.1f%%', startangle=90, colors=colors, pctdistance=0.75)
    axes[0].axis('equal')
    axes[0].set_title('FER2013 Augmented Train expression distribution')

    axes[1].pie(test_vals, labels=all_classes, autopct='%1.1f%%', startangle=90, colors=colors, pctdistance=0.75)
    axes[1].axis('equal')
    axes[1].set_title('FER2013 Augmented Test expression distribution')

    save_figure(fig, 'expression_distribution_train_vs_test_fer_aug.png')
    print('Saved combined augmented train vs test expression distribution plot.')
else:
    print('Skipping combined augmented train vs test distribution: datasets not loaded.')
```
